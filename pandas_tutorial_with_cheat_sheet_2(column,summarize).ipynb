### df.filter(regex =  ' ')
##### regex 는 정규식 표현으로 df를 필터링할 수 있는 파라미터다.
    '_'  : 칼럼에 _ 이 들어있는 칼럼을 골라낸다. 
    'length$':length로 끝나는 칼럼을 골라낸다 ($)
    '^sepal' : sepal로 시작하는 칼럼을 골라낸다(^). 
    '^(?!species$).*' : species가 포함된 컬럼 제외한 컬럼 //Matches strings except the string 'Species'
    '^x[1-5]$' :  x로 시작하면서 x다음에 오는 숫자 1부터 5까지오는 칼럼만 필터, // 
        Matches strings beginning with 'x' and ending with 1,2,3,4,5
    
### df.loc[2:7, 'sepal_length' : 'sepal_width'] (df.loc[row, column])
### df.iloc[2:7, [0,1]]  
    loc은 2:7에서 2<= <=7 , iloc은 2:7에서 2<= <7 이다.

### df['species'].value_counts()

### df.sum(), .min(), .max(), .std(), .var(), .counts(), .median(), .mean(), .quantile()

```python
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
columns = ['sepal_length', 'sepal_width', 'species']
df[columns].head()
```

```python
df.filter(regex = '_').head() # regex : 정규식을 넣어서 필터링할 수 있다
```

```python
df.filter(regex= 'length$').head() #  $ : length로 끝나는 단어인 컬럼 가져온다. 대소문자 구분해야함.
```

```python
df.filter(regex= '^sepal').head()  # ^ : sepal로 시작하는 컬럼을 가져온다.
```

```python
df.filter(regex= '^(?!species$).*').head() # ^(?!species$).* : species가 포함된 컬럼 제외한 컬럼 //Matches strings except the string 'Species'
```

```python
df.filter(regex = '^x[1-5]$').head() #x로 시작하면서 x다음에 오는 숫자 1부터 5까지오는 칼럼만 필터, Matches strings beginning with 'x' and ending with 1,2,3,4,5
```

```python
df.loc[2:7, 'sepal_length' : 'sepal_width'] #df.loc[row, column]
```

```python
df.iloc[2:7, [0,1]] #loc은 2:7에서 2<= <=7 , iloc은 2:7에서 2<= <7 이다.
```

```python
df.loc[df['sepal_length']>3, ['petal_width', 'sepal_width']].head()
```
